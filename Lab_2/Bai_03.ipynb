```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

```python
df = pd.read_csv('Data/pre_processed_data.csv', delimiter= ',')
df
```

Logistic regression

```python
df1= df.copy()
```

```python
def convert_log(Target : str) -> int:
    if Target == 'Graduate':
        return 1
    return 0
```

```python

df1['Target'] = df1['Target'].apply(convert_log)
X1 = df1.drop(columns ='Target')
Y1 = df1['Target']

X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1,Y1, train_size = 0.8, random_state= 24)
```

```python
scaler = StandardScaler()
X1_train = scaler.fit_transform(X1_train)
X1_test = scaler.transform(X1_test)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X1_train,Y1_train)
log_predicted = log_reg.predict(X1_test)
acc = metrics.accuracy_score(Y1_test, log_predicted)

print('Accuracy of logistic: ', acc)
```
